```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '1- mental-illnesses-prevalence.csv'  
data = pd.read_csv(file_path)

data.head()
```

```python
avg_depression = data.groupby('Year')["Depressive disorders (share of population) - Sex: Both - Age: Age-standardized"].mean()

plt.figure(figsize=(12, 6))
plt.plot(avg_depression, marker='o', label='Ortalama Depresyon Oranı', color='orange')
plt.title('Yıllara Göre Dünya Genelinde Ortalama Depresyon Oranı', fontsize=14)
plt.xlabel('Yıl', fontsize=12)
plt.ylabel('Oran (%)', fontsize=12)
plt.grid(alpha=0.5)
plt.legend()
plt.show()
```

```python
data_2010 = data[data['Year'] == 2010]

plt.figure(figsize=(14, 6))
sns.histplot(data_2010["Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized"], kde=True, bins=20, color='gold')
plt.title('2010 Yılında Ülkelerdeki Anksiyete Bozukluğu Dağılımı', fontsize=14)
plt.xlabel('Anksiyete Oranı (%)', fontsize=12)
plt.ylabel('Ülke Sayısı', fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```

```python
data_2010 = data[data['Year'] == 2010]

plt.figure(figsize=(14, 6))
sns.histplot(data_2010["Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized"], kde=True, bins=20, color='gold')
plt.title('2010 Yılında Ülkelerdeki Anksiyete Bozukluğu Dağılımı', fontsize=14)
plt.xlabel('Anksiyete Oranı (%)', fontsize=12)
plt.ylabel('Ülke Sayısı', fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```

```python
top_depression = data_2010.nlargest(5, "Depressive disorders (share of population) - Sex: Both - Age: Age-standardized")
bottom_depression = data_2010.nsmallest(5, "Depressive disorders (share of population) - Sex: Both - Age: Age-standardized")

print("En Yüksek Depresyon Oranına Sahip 5 Ülke (2010):")
print(top_depression[['Entity', 'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized']])

print("\nEn Düşük Depresyon Oranına Sahip 5 Ülke (2010):")
print(bottom_depression[['Entity', 'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized']])
```

```python
afghanistan_schizophrenia = data[data['Entity'] == 'Afghanistan'][['Year', "Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized"]]

plt.figure(figsize=(12, 6))
plt.plot(afghanistan_schizophrenia['Year'], afghanistan_schizophrenia["Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized"], marker='o', color='red')
plt.title('Afganistan\'da Yıllara Göre Şizofreni Oranı', fontsize=14)
plt.xlabel('Yıl', fontsize=12)
plt.ylabel('Oran (%)', fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```
